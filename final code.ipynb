```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
      L.append(i['name'])
  return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords']= movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```
